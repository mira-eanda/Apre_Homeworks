```python
import numpy as np

#observations
x00 = np.array([[1], [1], [1], [1]])
x01 = np.array([[1], [0], [0], [-1]])

#targets
t_x0 = np.array([[-1], [1], [-1]])
t_x1 = np.array([[1], [-1], [-1]])

#weights
w1 = np.array([[1, 1, 1, 1], [1, 1, 2, 1], [1, 1, 1, 1]])
w2 = np.array([[1, 4, 1], [1, 1, 1]])
w3 = np.array([[1, 1], [3, 1], [1, 1]])

#bias
b1 = np.array([[1], [1], [1]])
b2 = np.array([[1], [1]])
b3 = np.array([[1], [1], [1]])

#activation function
# f(x) = ùë°ùëéùëõ‚Ñé(0.5ùë•‚àí2)
def f(x):
	return np.tanh(0.5*x - 2)

def deriv_f(x):
	d = 2 / (np.exp(2-0.5*x)+np.exp(0.5*x-2)) **2
	return d
```

```python
# ----------------------------Forward Pass for x0

z1_0 = np.dot(w1, x00) + b1
# print("z1_0: \n", z1_0)

x1_0 = f(z1_0)
# print("x1_0: \n", x1_0)

z2_0 = np.dot(w2, x1_0) + b2
# print("z2_0: \n", z2_0)

x2_0 = f(z2_0)
# print("x2_0: \n", x2_0)

z3_0 = np.dot(w3, x2_0) + b3
# print("z3_0: \n", z3_0)

x3_0 = f(z3_0)
# print("x3_0: \n", x3_0)

# -------------------------- Forward Pass for x1
z1_1 = np.dot(w1, x01) + b1
# print("z1_1: \n", z1_1)

x1_1 = f(z1_1)
# print("x1_1: \n", x1_1)

z2_1 = np.dot(w2, x1_1) + b2
# print("z2_1: \n", z2_1)

x2_1 = f(z2_1)
# print("x2_1: \n", x2_1)

z3_1 = np.dot(w3, x2_1) + b3
# print("z3_1: \n", z3_1)

x3_1 = f(z3_1)
# print("x3_1: \n", x3_1)

# -------------------------- Backward Pass

# deltas

delta3_0 = (x3_0 - t_x0) * deriv_f(z3_0)
# print("delta3_0: \n", delta3_0)

delta2_0 = np.dot(np.transpose(w3), delta3_0) * deriv_f(z2_0)
# print("delta2_0: \n", delta2_0)

delta1_0 = np.dot(np.transpose(w2), delta2_0) * deriv_f(z1_0)
# print("delta1_0: \n", delta1_0)

delta3_1 = (x3_1 - t_x1) * deriv_f(z3_1)
# print("delta3_1: \n", delta3_1)

delta2_1 = np.dot(np.transpose(w3), delta3_1) * deriv_f(z2_1)
# print("delta2_1: \n", delta2_1)

delta1_1 = np.dot(np.transpose(w2), delta2_1) * deriv_f(z1_1)
# print("delta1_1: \n", delta1_1)

# gradients and updated weights and biases

dEdW3 = delta3_0 * np.transpose(x2_0) + delta3_1 * np.transpose(x2_1)
# print("dEdW3: \n", dEdW3)

new_w3 = w3 - 0.1 * dEdW3
# print("new_w3: \n", new_w3)

new_b3 = b3 - 0.1 * (delta3_0 + delta3_1) # b3 - 0.1 *dEdB3
# print("new_b3: \n", new_b3)


dEdW2 = delta2_0 * np.transpose(x1_0) + delta2_1 * np.transpose(x1_1)
# print("dEdW2: \n", dEdW2)

new_w2 = w2 - 0.1 * dEdW2
# print("new_w2: \n", new_w2)

new_b2 = b2 - 0.1 * (delta2_0 + delta2_1)
# print("new_b2: \n", new_b2)


dEdW1 = delta1_0 * np.transpose(x00) + delta1_1 * np.transpose(x01)
# print("dEdW1: \n", dEdW1)

new_w1 = w1 - 0.1 * dEdW1
# print("new_w1: \n", new_w1)

new_b1 = b1 - 0.1 * (delta1_0 + delta1_1)
# print("new_b1: \n", new_b1)







```
